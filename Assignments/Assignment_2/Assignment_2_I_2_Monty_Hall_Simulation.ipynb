```python
"""
This file contains code for Monty Hall Simulation
Author : Sanjay Marreddi
Date   : 14-10-2020
"""

# Importing the required Libraries
import random
from numpy import mean, cumsum
from matplotlib import pyplot as plt
```

```python
# Defining a function which does Monty Hall Simulation.
def simulate(make_switch=False, n_doors=4):
 
    # sample one index for the door to hide the car behind
    prize_index = random.randint(0, n_doors - 1)

    # sample one index for the door initially chosen by the player
    choice_index = random.randint(0, n_doors - 1)

    # we can test for the current result
    current_result = prize_index == choice_index

    # now Monty Hall opens all doors the player did not choose, except for one door
    if (make_switch):
        # if we do switch, we change to the one remaining door, which inverts our current choice
        return not current_result
    else:
        # the player sticks with his/her original door,
        return current_result
```

```python
# Choosing the number of games we want to simulate
n_games = 1000

# simulate the games and store the boolean results
results_with_switching = [simulate(make_switch=True) for _ in range(n_games)]
results_without_switching = [simulate(make_switch=False) for _ in range(n_games)]

# make a equal-length list showing, for each element in the results, the game to which it belongs
games = [i + 1 for i in range(n_games)]

# make a line plot of the cumulative wins with and without switching
plt.figure(figsize=(7, 7))
plt.plot(games, cumsum(results_with_switching), color='green', label='switching')
plt.plot(games, cumsum(results_without_switching), color='blue', label='no switching')
plt.axis([0, n_games, 0, n_games])
plt.legend()
plt.xlabel('Number of games played')
plt.ylabel('Cumulative number of games won by Contestant')
plt.show()

print(f'Switching doors makes the Contestant win {sum(results_with_switching)} of {n_games} games ({mean(results_with_switching) * 100:.1f}%)'
    f'\nWhere as when not switching Contestant wins only {sum(results_without_switching)} games ({mean(results_without_switching) * 100:.1f}%)')
```
