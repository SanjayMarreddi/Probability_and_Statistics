```python
"""
This file contains code for plotting the probability mass function 
and cumulative distribution function of the Binomial and Geometric Random Variables.

Author : Sanjay Marreddi
Date   : 14-10-2020
"""

# Importing the required Libraries
import numpy as np
from scipy.stats import binom 
import matplotlib.pyplot as plt 

%matplotlib inline
```

## Binomial Random Variable

### 4(a) Binomial, n = 30, p = 0.2

#### Probability Mass Function (PMF)

```python
n=30
p=0.2
rv = binom(n,p)
x=np.arange(0,n+1,1)
pmf = rv.pmf(x)

plt.figure(figsize=(12,6))
plt.title("Probability mass function: $\\binom{n}{k}\, p^k (1-p)^{n-k}$\n",fontsize=20)
plt.scatter(x,pmf)
plt.legend(["$n=30, p=0.2$"],fontsize=15)
plt.xlabel("Number of successful trials ($k$)",fontsize=15)
plt.ylabel("Probability of success",fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.grid(True)
plt.show()
```

### 4(a) Binomial, n = 30, p = 0.2

#### Cumulative Distribution Function (CDF)

```python
n=30
p=0.2
rv = binom(n,p)
x=np.arange(0,31,1)
cdf = rv.cdf(x)

plt.figure(figsize=(12,6))
plt.title("Cumulative distribution function\n",fontsize=20)
plt.scatter(x,cdf)
plt.legend(["$n=30, p=0.2$"],fontsize=15)
plt.xlabel("Number of successful trials",fontsize=15)
plt.ylabel("Cumulative probability of success",fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.grid(True)
plt.show()
```

## Geometric Random Variable

### 4(b) Geometric, k = 1 . . . , 20, p = 0.2

#### Probability Mass Function (PMF) 

```python
# As the formulae for PMF and CDF for Geometric Random Variable are Small, Lets Implement without using Scipy
```

```python
p = 0.2
n= 20
```

```python
xrange = list(range(1,n))
pmf = []
cdf =[]
cdf_i =0
for i in range(1,n):
    
    # Evaluating the PMF using Definition
    pmf_i = (p)* ((1-p)**(i-1))
    pmf.append(pmf_i)
    
    cdf_i+= pmf_i
    cdf.append(cdf_i)
```

```python
plt.figure(figsize=(12,6))
plt.title(" Probability Mass Function\n",fontsize=20)
plt.plot(xrange,pmf)
plt.legend(["$p=0.2$"],fontsize=15)
plt.xlabel("Value of k",fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.grid(True)
plt.show()
```

#### Cumulative Distribution Function (CDF)

```python
plt.figure(figsize=(12,6))
plt.title(" Cumulative Distribution Function\n",fontsize=20)
plt.plot(xrange,cdf)
plt.legend(["$p=0.2$"],fontsize=15)
plt.xlabel("Value of k",fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.grid(True)
plt.show()
```
